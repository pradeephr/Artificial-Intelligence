Optimal MDP Policy using Value Iteration

Determine the travel path using value iteration 

```python
#The following variable must be used in the You can change the location values for testing purposes
stoneLocation=[2, 8]  #[row, column]
positiveTerminalLocation = [0,9,2]  #[row, column, utility value]
nagativeTerminalLocation = [1,9, -2]  #[row, column, utility value]
iteration=20
noise=0.15
discount=0.91
n=10
r=n
c=n
```

```python
import enum
from copy import deepcopy

class d(enum.Enum):
    E='e'#east
    W='w'#west
    N='n'#north
    S='s'#south


a=[[0 for i in range(c)] for j in range(r)]
a[stoneLocation[0]][stoneLocation[1]]='B'
a[positiveTerminalLocation[0]][positiveTerminalLocation[1]]=positiveTerminalLocation[2]
a[nagativeTerminalLocation[0]][nagativeTerminalLocation[1]]=nagativeTerminalLocation[2]


aval=[[0 for i in range(c)] for j in range(r)]
atemp=[[0 for i in range(c)] for j in range(r)]
adir=[['n' for i in range(c)] for j in range(r)]


max=0
min=-1
p1=(1-noise)
s1=(noise/2)
    

def init_vals():
    global r
    global c
    global max
    global min
    global a
    global aval
    global atemp
    for i in range(r):
        for j in range(c):
            if(a[i][j]=='B'):
                aval[i][j]=0
                atemp[i][j]=0
            else:
                aval[i][j]=a[i][j]
                atemp[i][j]=a[i][j]
                if(a[i][j]<0):
                    min=a[i][j]
                elif(a[i][j]>0):
                    max=a[i][j]             
    
def update_policy():
    global r
    global c
    global atemp
    global aval
    global a
    global d
    global p1
    global s1
    global min
    global max
    j=0
    
    for i in range(r):
       for j in range(c):
           if not (a[i][j]=='B' or a[i][j]==max or a[i][j]==min):
               eVal=((aval[i][j]*s1) if (i==0 or a[i][j]=='B') else (aval[i-1][j]*s1))
               eVal+=((aval[i][j]*s1) if (i==r-1 or a[i+1][j]=='B') else (aval[i+1][j]*s1))
               eVal+=((aval[i][j]*p1) if (j==c-1 or a[i][j+1]=='B') else (aval[i][j+1]*p1)) 
               wVal=((aval[i][j]*s1) if (i==0 or a[i][j]=='B') else (aval[i-1][j]*s1))
               wVal+=((aval[i][j]*s1) if (i==r-1 or a[i+1][j]=='B') else (aval[i+1][j]*s1))
               wVal+=((aval[i][j]*p1) if (j==0 or a[i][j-1]=='B') else (aval[i][j-1]*p1)) 
               nVal=((aval[i][j]*s1) if (j==0 or a[i][j-1]=='B') else (aval[i][j-1]*s1))
               nVal+=((aval[i][j]*s1) if (j==c-1 or a[i][j+1]=='B') else (aval[i][j+1]*s1))
               nVal+=((aval[i][j]*p1) if (i==0 or a[i-1][j]=='B') else (aval[i-1][j]*p1)) 
               sVal=((aval[i][j]*s1) if (j==0 or a[i][j-1]=='B') else (aval[i][j-1]*s1))
               sVal+=((aval[i][j]*s1) if (j==c-1 or a[i][j+1]=='B') else (aval[i][j+1]*s1))
               sVal+=((aval[i][j]*p1) if (i==r-1 or a[i+1][j]=='B') else (aval[i+1][j]*p1)) 
               eVal=round(eVal*discount,2)
               wVal=round(wVal*discount,2)
               nVal=round(nVal*discount,2)
               sVal=round(sVal*discount,2)
               
               atemp[i][j]=eVal
               adir[i][j]='e'
               if(atemp[i][j]<wVal):
                  atemp[i][j]=wVal
                  adir[i][j]='w'
               if(atemp[i][j]<nVal):
                  atemp[i][j]=nVal
                  adir[i][j]='n'
               if(atemp[i][j]<sVal):
                  atemp[i][j]=sVal
                  adir[i][j]='s'
    aval=deepcopy(atemp)

def print_result(aval,adir):
    
    for i in range(r):
        for j in range(c):
            if(i==stoneLocation[0] and j==stoneLocation[1]):
                print('STONE',end=',');
            else:
                p=','
                if(j==c-1):
                    p='\n'    
                D=adir[i][j]
                res=str(aval[i][j])+'('+D+')';
                print(res,end=p)
        print()
    print("----------------------------------------\n") 
    
init_vals()    

def start():
    global aval
    global adir
    for i in range(iteration):
        print("Values for iteration ",i+1)
        print_result(aval,adir)
        update_policy()
      
```
