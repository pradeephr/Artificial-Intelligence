Bot1 Vs Bot2 using MinMax tree and AlphaBeta pruning

```python
from random import randint;

class node:
    a=None
    score=None
    parent=None
    children=None
    
    def __init__(self,arr):
        self.a=list(arr)
        #self.getScore()
            
    def gameEnd(self):
        for i in range(9):
            if(self.a[i]==0):
                return False
        return True
    
     
        
    def getChildren(self,t):
        children=list()
        for i in range(9):
            if(self.a[i]==0):
                child=node(self.a)
                child.a[i]=t
                children.append(child)
        return children        
def getScore(node,d):
        score=0
        #print("calculating score for child ",node.a)
        if((node.a[0]==node.a[4]==node.a[8]==1) or 
           (node.a[2]==node.a[4]==node.a[6]==1) or
            (node.a[0]==node.a[1]==node.a[2]==1) or
             (node.a[3]==node.a[4]==node.a[5]==1) or
              (node.a[6]==node.a[7]==node.a[8]==1) or
               (node.a[0]==node.a[3]==node.a[6]==1) or
                (node.a[1]==node.a[4]==node.a[7]==1) or
                 (node.a[2]==node.a[5]==node.a[8]==1)):
                     score=10
        elif((node.a[0]==node.a[4]==node.a[8]==2) or 
           (node.a[2]==node.a[4]==node.a[6]==2) or
            (node.a[0]==node.a[1]==node.a[2]==2) or
             (node.a[3]==node.a[4]==node.a[5]==2) or
              (node.a[6]==node.a[7]==node.a[8]==2) or
               (node.a[0]==node.a[3]==node.a[6]==2) or
                (node.a[1]==node.a[4]==node.a[7]==2) or
                 (node.a[2]==node.a[5]==node.a[8]==2)):
                     score=-10
        else:
            score=0
        #print("score is ",score)    
        return score   

def minmax(node,t,d):
    global nextMove
    if(node.gameEnd()):
        #print(node.a)
        score=getScore(node,d)
        #print(score)
        return score
    elif(t==1):
        children=node.getChildren(t)
        maxScore=-99
        for child in children:
            val=minmax(child,2,d+1)
            if(val>maxScore):
                maxScore=val
        return maxScore    
    elif(t==2):
        children=node.getChildren(t)
        minScore=99
        for child in children:
            val=minmax(child,1,d+1)            
            if(val<minScore):
                minScore=val
        return minScore        

def findBestMaxMove(root,d):
    bestScore=-99
    bestMove=node(root.a)  
    #print("inside max",root.a)
    for i in range(9):
        if(root.a[i]==0):
            root.a[i]=2
            if(getScore(root,d)==-10):
                root.a[i]=1
                bestMove=node(root.a)
                #print("returing ",root.a)
                return bestMove
            root.a[i]=0    
    for i in range(9):
        if(root.a[i]==0):
            root.a[i]=1  
            score=minmax(root,2,d)
            #print("max score for move", root.a,":",score)
            root.a[i]=0
            if(score>bestScore):
                bestMove=node(root.a)
                bestMove.a[i]=1
                bestScore=score
    return bestMove            

def findBestMinMove(root,d):
    minBestScore=99 
    minBestMove=node(root.a)
    #print("inside min",root.a)
    for i in range(9):
        if(root.a[i]==0):
            root.a[i]=1
            #print("score is ",getScore(root,d))
            if(getScore(root,d)==10):
                root.a[i]=2
                minBestMove=node(root.a)
                minBestScore=-10
                return minBestMove
            root.a[i]=0
            
    for i in range(9):
        if(root.a[i]==0):
            root.a[i]=2
            score=minmax(root,1,d)
            #print("min score for move", root.a,":",score)
            root.a[i]=0
            if(score<minBestScore):
                minBestMove=node(root.a)
                minBestMove.a[i]=2
                minBestScore=score
    return minBestMove
              



a=[0 for i in range(9)]
r=randint(0,8)
a[r]=1
root=node(a)
t=1
val=-1
d=0

def print_result(arr):
    for i in range(3):
        for j in range(3):
            if(arr[(i*3)+j]==0):
                print("*",end='')
            elif(arr[(i*3)+j]==1):
                print("X",end='')
            elif(arr[(i*3)+j]==2):
                print("O",end='')
        print()
    print("------------------")    
                
def start():
    global root
    global t
    global d
    print_result(root.a)
    while(root.score!=10 and root.score!=-10):
        flag=0
        for i in range(9):
            if(root.a[i]==0):
                flag=1
        if(flag==0):
            break
        bestMove=findBestMinMove(root,d)
        root=node(bestMove.a)
        #print(root.a)
        print_result(bestMove.a)
        t=1 if t==2 else 2
        bestMove=findBestMaxMove(root,d)
        root=node(bestMove.a)
        #print(root.a)
        print_result(bestMove.a)
        d+=1    
```
